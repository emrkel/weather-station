```python
import pandas as pd
import matplotlib.pyplot as plt

pd.options.display.float_format = '{:.4f}'.format
```

```python
# Read the data from the database
query = "SELECT * FROM archive"
allData = pd.read_sql(query, connection)

allData.head(10)
```

```python
from datetime import datetime

allData['dateTime'] = allData['dateTime'].apply(lambda x: datetime.fromtimestamp(x))
allData.set_index('dateTime', inplace=True)
allData
```

```python
# Convert the temperatures from Fahrenheit to Celsius

def celcius_convert(f):
    if f is not None:
        return ((f-32)*5)/9
    else:
        return None


temps = ['inTemp', 'outTemp', 'extraTemp1', 'extraTemp2', 'extraTemp3', 'soilTemp1', 'soilTemp2', 'soilTemp3', 'soilTemp4', 'leafTemp1', 'leafTemp2', 'heatingTemp', 'appTemp', 'windchill', 'dewpoint', 'heatindex']


for temp in temps:
	allData[temp] = allData[temp].apply(lambda x: celcius_convert(x))
 
allData.tail(10)
```

```python
query = "SELECT * FROM archive_day_outTemp"
outside_temp = pd.read_sql(query, connection)

outside_temp.head(10)

outside_temp['dateTime'] = outside_temp['dateTime'].apply(lambda x: datetime.fromtimestamp(x))
outside_temp['mintime'] = outside_temp['mintime'].apply(lambda x: datetime.fromtimestamp(x))
outside_temp['maxtime'] = outside_temp['maxtime'].apply(lambda x: datetime.fromtimestamp(x))

outside_temp['min'] = outside_temp['min'].apply(lambda x: celcius_convert(x))
outside_temp['max'] = outside_temp['max'].apply(lambda x: celcius_convert(x))
outside_temp
```
