```python
# Read the data from the database - Takes 53 seconds on average to read 1.1 million rows
query = "SELECT * FROM archive"
allData = pd.read_sql(query, connection)

allData.head(10)
```

```python
from datetime import datetime

allData['dateTime'] = allData['dateTime'].apply(lambda x: datetime.fromtimestamp(x))
allData.set_index('dateTime', inplace=True)
allData
```

```python
# Convert the temperatures from Fahrenheit to Celsius

def celcius_convert(f):
    if f is not None:
        return ((f-32)*5)/9
    else:
        return None


temps = ['inTemp', 'outTemp', 'extraTemp1', 'extraTemp2', 'extraTemp3', 'soilTemp1', 'soilTemp2', 'soilTemp3', 'soilTemp4', 'leafTemp1', 'leafTemp2', 'heatingTemp', 'appTemp', 'windchill', 'dewpoint', 'heatindex']


for temp in temps:
	allData[temp] = allData[temp].apply(lambda x: celcius_convert(x))
 
 
# Convert the wind speed/gust from mph to m/s

def wind_convert(mph):
    if mph is not None:
        return mph * 0.44704
    else:
        return None

allData['windSpeed'] = allData['windSpeed'].apply(lambda x: wind_convert(x))
allData['windGust'] = allData['windGust'].apply(lambda x: wind_convert(x))

# Convert rain from inches to mm

def rain_convert(inches):
    if inches is not None:
        return inches * 25.4
    else:
        return None
    
def rainrate_convert(iph):
    if iph is not None:
        return iph * 25.4
    else:
        return None
    
allData['rain'] = allData['rain'].apply(lambda x: rain_convert(x))
allData['rainRate'] = allData['rainRate'].apply(lambda x: rainrate_convert(x))
    
# Convert pressure, alimeter and barometer from inHg (inches of Mercury) to mbar (millibar)

def pressure_convert(inHg):
    if inHg is not None:
        return inHg * 33.8639
    else:
        return None
    
allData['pressure'] = allData['pressure'].apply(lambda x: pressure_convert(x))
allData['altimeter'] = allData['altimeter'].apply(lambda x: pressure_convert(x))
allData['barometer'] = allData['barometer'].apply(lambda x: pressure_convert(x))
    
 
allData.tail(10)
```

```python
# Key numbers 

# Fetch the inside temperature (inTemp)
def get_in_temp():
	cursor.execute("SELECT inTemp FROM archive ORDER BY dateTime DESC LIMIT 1")
	temp = cursor.fetchone()
	farenheit = float(temp[0])
	inTemp = celcius_convert(farenheit)
	return round(inTemp, 2)

# Fetch the outside temperature (outTemp)
def get_out_temp():
	cursor.execute("SELECT outTemp FROM archive ORDER BY dateTime DESC LIMIT 1")
	temp = cursor.fetchone()
	farenheit = float(temp[0])
	outTemp = celcius_convert(farenheit)
	return round(outTemp, 2)

def get_rainfall():
    query = "SELECT dateTime, rain FROM archive"
    rain = pd.read_sql(query, connection)
    rain['dateTime'] = rain['dateTime'].apply(lambda x: datetime.fromtimestamp(x))
    rain.set_index('dateTime', inplace=True)
    rain['rain'] = rain['rain'].apply(lambda x: rain_convert(x))
    rain = rain.resample('D').sum()
    return round(rain.iloc[-1]['rain'], 2)

def get_humidity():
	cursor.execute("SELECT outHumidity FROM archive ORDER BY dateTime DESC LIMIT 1")
	humidity = cursor.fetchone()
	return humidity[0]

```
